```python
top_roms = df.sort_values(by='ROMS', ascending=False).head(10)
bottom_roms = df.sort_values(by='ROMS').head(10)
top_roms
bottom_roms
```

```python
df['Efficient'] = df['ROMS'] > 1
df['Efficient'].value_counts()
```

```python
top_5 = df.sort_values(by='ROMS', ascending=False).head(5)
top_5[['Store ID', 'City', 'State', 'Sales Region', 'New Expansion', 'Marketing Spend', 'Revenue', 'Profit', 'ROMS']]
```

```python
rest = df[~df['Store ID'].isin(top_5['Store ID'])]

comparison = pd.DataFrame({
    'Top 5 Average': top_5[['Marketing Spend', 'Revenue', 'Profit', 'ROMS']].mean(),
    'Others Average': rest[['Marketing Spend', 'Revenue', 'Profit', 'ROMS']].mean()
})
comparison
```

```python
top_5['Sales Region'].value_counts()
top_5['New Expansion'].value_counts()
```

```python
comparison.plot(kind='bar', figsize=(8,5))
plt.title('Top 5 Stores vs Others')
plt.ylabel('Value')
plt.xticks(rotation=0)
plt.show()
```

```python
sns.histplot(df['ROMS'], bins=20, kde=True)
plt.title('ROMS Distribution')
plt.show()
```

```python
sns.lmplot(x='Marketing Spend', y='ROMS', hue='Sales Region', data=df)
plt.title('Marketing Spend vs ROMS by Region')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Marketing Spend', y='Profit', data=df)
plt.title('Marketing Spend vs Profit')
plt.xlabel('Marketing Spend')
plt.ylabel('Profit')
plt.show()
```

```python
df.groupby('Sales Region')[['Marketing Spend', 'Revenue', 'Profit']].mean().plot(kind='bar')
plt.title('Average Metrics by Sales Region')
plt.xticks(rotation=0)
plt.ylabel('Amount')
plt.show()
```

```python
sns.boxplot(x='Sales Region', y='Profit', data=df)
plt.title('Profit Distribution by Sales Region')
plt.show()
```
